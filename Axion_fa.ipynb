# Constraints on $1/f_a$ scale versus axion mass

QCD axion band defined by
\begin{equation}
m_{a}=5.70(7) \mu \mathrm{eV}\left(\frac{10^{12} \,\mathrm{GeV}}{f_{a}}\right)
\end{equation}

```python
from numpy import *
import matplotlib.pyplot as plt
from PlotFuncs import BlackHoleSpins,Axion_fa,MySaveFig,FigSetup
    
fig,ax = FigSetup(Shape='Rectangular',ylab=r'$1/f_a$ [GeV$^{-1}$]',mathpazo=True,\
                              g_min=1e-20,g_max=1e-7,m_min=1e-20,m_max=1e-2)

BlackHoleSpins(ax,1,[4e-11,0.4e-17],fs=27,rotation=-90,text_col='w',alpha=1,facecolor='gray',text_on=True)
BlackHoleSpins(ax,1,[4e-11*(1+0.01),0.4e-17*(1+0.02)],fs=27,whichfile='Baryakhtar',rotation=-90,text_col='k',alpha=1,facecolor='darkgray',text_on=True)

Axion_fa.QCDAxion(ax)
Axion_fa.nEDM(ax)
Axion_fa.SolarCore(ax)
Axion_fa.Pulsars(ax)
Axion_fa.SN1987A(ax)
Axion_fa.GW170817(ax)
Axion_fa.BBN(ax)
    
plt.gcf().text(0.5,0.31,'Mehta et al.',fontsize=20,rotation=90)
plt.gcf().text(0.465,0.17,'Baryakhtar et al.',fontsize=20,rotation=90)

plt.text(9e-17,1.7e-19,'SMBHs')
plt.plot([1.2e-17,8e-17],[1e-19,2e-19],'k-',lw=2)

plt.text(2.5e-10,0.5e-13,'Stellar BHs')
plt.plot([1.5e-11,2e-10],[1e-14,5e-14],'k-',lw=2)

MySaveFig(fig,'Axion_fa')
```
